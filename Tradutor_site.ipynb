```python
import requests
from bs4 import BeautifulSoup

def extract_text_from_url(url):
    response = requests.get(url)
    part = ''

    if response.status_code == 200:
        soup = BeautifulSoup(response.tex, 'html.parser')
        for script_or_style in soup(['script','style']):
            script_or_style.decompose()
        texto = soup.get_text(separator = ' ')
        #Limpar texto
        linhas = (line.strip() for line in texto.splitlines())
        parts = (phrase.strip() for line in linhas for phrase in line.split(" "))
        texto_limpo = ' '.join(part in parts if part)
        return texto_limpo
        
    else:
        print("Failed to fetch the URL. Status code: {response.status_code}")
        return None
    
    text = soup.get_text()
    return text

extract_text_from_url('https://dev.to/kenakamu/azure-open-ai-in-vnet-3alo')
```
